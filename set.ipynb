```python
m=set(["monday","tuesday","wednesday","thursday"])
print(m)
m.add("sunday")
print(m)
```

```python
m=set(["monday","tuesday","wednesday","thursday"])
print(m)
m.update(["saturday","sunday"])
print(m)
```

```python
m=set(["monday","tuesday","wednesday","thursday"])
m.discard("monday")
print(m)
m.remove("tuesday")
print(m)
m.pop()
print(m)
m.clear()
print(m)
m.discard("monday")
print(m)
m.remove("monday")
```

```python
m1={"a","b","c","d"}
m2={"a","b","e","f"}
m3={"a","z"}
print(m1|m2)
print(m1.union(m2))
print(m1.union(m2,m3))
```

```python
m1={"a","b","c"}
m2={"a","b","e","c"}
m3={"a","r","q","c"}
m=m1.intersection(m2,m3)
print(m)
m1.intersection_update(m2,m3)
print(m1)
print(m1&m2&m3)
```

```python
m1={"a","b","c","d"}
m2={"a","b","c","r"}
print(m1-m2)
print(m1.difference(m2))
m11=m1^m2
print(m11)
m111=m1.symmetric_difference(m2)
print(m111)
print(m1>m2)
print(m1<m2)
print(m1==m2)
```
