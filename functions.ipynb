```python
def function1( *args_list ):
  ans = []
  for l in args_list:
    ans.append(l.upper())
  return ans
def function2( **kargs_list ):
  ans = []
  for key, value in kargs_list.items():
    ans.append([key, value])
  return ans
obj=function1('java', 'lessons', 'beginners')
print(obj)
obj2=function2(First='java',Second='lessons',Third='beginners')
print(obj2)
```

```python
def fact(num):
  if num == 1:
    return 1
  else:
    return (num * fact(num-1))
num = int(input("Enter a number: "))
result = fact(num)
print(num,"factorial:", result)
```
